```python
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
df.info()
```

```python
X = df[['Years_of_Experience', 'Mental_Health_Condition', 'Productivity_Change', 'Social_Isolation_Rating', 'Satisfaction_with_Remote_Work', 'Sleep_Quality', 'Physical_Activity']]
y = df['Stress_Level']
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.1, random_state = 45)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.1, random_state = 45)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_val = scaler.fit_transform(X_val)
```

```python
ann_model = Sequential()
```

```python
#!pip install tensorflow
```

```python
from tensorflow.keras import regularizers

ann_model.add(Dense(4, input_dim=X_train.shape[1], activation='relu', kernel_regularizer = regularizers.l2(0.001)))

ann_model.add(Dense(64, activation = 'relu', kernel_regularizer = regularizers.l2(0.001)))

ann_model.add(Dense(128, activation = 'relu', kernel_regularizer = regularizers.l2(0.001)))
ann_model.add(Dropout(0.5)) 

ann_model.add(Dense(32, activation = 'relu', kernel_regularizer = regularizers.l2(0.001)))

ann_model.add(Dense(1, activation = 'linear'))

ann_model.compile(optimizer = 'adam', loss = 'mean_squared_error')

hist = ann_model.fit(X_train, y_train, epochs=50, validation_data = (X_val, y_val), batch_size=32)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = ann_model.predict(X_test)

ann_mae = mean_absolute_error(y_test, y_pred)
ann_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
ann_r2 = r2_score(y_test, y_pred)

print(f"ANN MAE: {ann_mae}")
print(f"ANN RMSE: {ann_rmse}")
print(f"ANN R2: {ann_r2}")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, f1_score, recall_score
```

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 45)

rf_model.fit(X_train, y_train)

cv_rf = cross_val_score(rf_model, X_train, y_train, cv = 10, scoring = 'accuracy')

y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)

rf = cv_rf.mean()
print(f'Mean CV Accuracy: {cv_rf}')
```
